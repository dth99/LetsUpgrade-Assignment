### Visualization

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('diabetes.csv',index_col=0)
```

```python
data.columns
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
sns.countplot(x='Age',data=df)
```

```python
sns.countplot(x='Outcome',data=df)
```

```python
sns.countplot(x='BloodPressure',data=df)
```

### stat in ratio

```python
sns.countplot(x='Age',hue='Outcome',data=df)
```

```python

sns.relplot(x='Age',y='BloodPressure',hue='Outcome',data=df)
```

```python
sns.relplot(x='Age',y='Pregnancies',hue='Outcome',data=df)
```

```python
sns.catplot(x='Age',y='BMI',hue='Outcome',data=df)
```
